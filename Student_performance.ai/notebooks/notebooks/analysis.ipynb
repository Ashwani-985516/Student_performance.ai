```python
df['Percentage'] = (df['ExamScore'] / df['TotalExamMarks']) * 100
df['Percentage'] 
```

```python
df['AttendancePercentage'] = (df['AttendedClasses'] / df['TotalClasses']) * 100
df['AttendancePercentage']
```

```python
df['Percentage'] = (df['ExamScore'] / df['TotalExamMarks']) * 100
df['Percentage']
df['Outcome'] = (df['Percentage'] < 50).astype(int)
df['Outcome']
```

```python
df['AttendancePercentage'] = (df['AttendedClasses'] / df['TotalClasses']) * 100
df['AttendancePercentage']
```

```python
pass_students = df[df['Outcome'] == 0]

pass_students[['Name',
               'TotalClasses',
               'AttendedClasses',
               'AttendancePercentage',
               'StudyHours',
               'Assignments',
               'StressLevel',
               'ExamScore']]

pass_students.to_excel('pass_students.xlsx', index=False)
```

```python
fail_students = df[df['Outcome'] == 1]

fail_students[['Name',
               'TotalClasses',
               'AttendedClasses',
               'AttendancePercentage',
               'StudyHours',
               'Assignments',
               'StressLevel',
               'ExamScore']]

fail_students.to_excel('fail_students.xlsx', index=False)
```
